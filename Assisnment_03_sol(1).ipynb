```python
import numpy as np
```

```python
# material
m = 1.0
k = 4.0

# frequence
wn = pow(k/m, 0.5)
Tn = 2.0*np.pi/wn

# damping
xi = 0.05
c = 2.0*m*xi*wn
wD = pow(1-xi*xi, 0.5)*wn

# I.C
uo = 1.0
vo = 1.0

# time
dt = 1e-4
if (dt > 0.1*Tn):
    print("dt = Tn/10")
    dt = 0.1*Tn
EndTime = 2.0*Tn+dt/10
```

```python
A = uo
B = (vo+xi*wn*uo)/wD
phi = np.arctan(B/A)
ome = np.arctan(wD/(xi*wn))
def func_u(t):
    return pow(A*A+B*B, 0.5)*np.exp(-xi*wn*t)*np.cos(wD*t-phi)  
```

```python
tns = np.arange(0, EndTime, dt)
len_t = len(tns)
un = np.zeros(len_t)
for i in range(len_t):
    un[i] = func_u(tns[i])
MAX_UN = max(np.abs(un))
print("Max|u(t)| = %6.4f\n"%MAX_UN)

max_t = (-np.pi/2 + phi+ ome)/wD
print("t = %6.4f"%max_t)
print("max|u(t)| = %6.4f"%func_u(max_t))
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(16, 9))
plt.plot(tns, un, lw=5, color="tab:blue", label="u(t)")
plt.xlim(0, EndTime)
plt.xlabel("time (s)", fontsize=30)
plt.ylabel("u(t) (m)", fontsize=30)
plt.legend(fontsize=30)
plt.grid(True)
plt.show()
```
